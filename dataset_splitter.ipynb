```
import os
from glob import glob
import shutil
from sklearn.model_selection import train_test_split
```

```
def the_splitter(train_size, val_size, test_size, the_path):
  image_files = the_path
  image_names_x = [name.replace(".jpg","")for name in image_files]
  image_names = [name.replace("drive/MyDrive/thesis/","") for name in image_names_x]
  train_x_names, test_names = train_test_split(image_names, test_size=test_size, train_size=train_size)
  train_names, val_names = train_test_split(train_x_names, test_size=val_size)
  d = dict()
  d['train']=train_names
  d['val']=val_names
  d['test']=test_names
  # x_count=len(train_names)+len(val_names)+len(test_names)
  print(image_names)
  # print(train_size)
  # print(val_size)
  # print(test_size)
  # print(x_count)
  # print(len(train_names))
  # print(len(train_x_names))
  # print(len(test_names))
  # print(len(val_names))
  return d
```

```
# NotebookApp.iopub_data_rate_limit=1000000000
# do test train splitting

test_size=0.2
train_size=0.8
val_size=train_size/6.4
image_files=[]

# find image names
the_classes=['ambulance','bus','car','taxi','truck','van','vehicle registration plate']

for the_class in the_classes:
  image_file= glob("drive/MyDrive/thesis/OID/"+the_class+"/images/*.jpg")
  image_files = image_files+image_file

               

x_dataset=the_splitter(train_size, test_size, val_size, image_files)
```

```
# print(len(image_names))
train_names =x_dataset['train']
val_names = x_dataset['val']
test_names = x_dataset['test']
print(len(train_names))
print(len(val_names))
print(len(test_names))
```

```
# print(len(image_names))
print(len(x_dataset['train']))
print(len(x_dataset['val']))
print(len(x_dataset['test']))
```

```
valNames_w=val_names
vallabel_names = [name.replace("images","pascal") for name in valNames_w]

trainNames_w=train_names
trainlabel_names = [name.replace("images","pascal") for name in trainNames_w]

testNames_w=test_names
testlabel_names = [name.replace("images","pascal") for name in testNames_w]
```

```
with open("train.txt", "w") as output:
    for s in trainlabel_names:
        output.write(str(s) + '\n')

with open("val.txt", "w") as output:
    for s in vallabel_names:
        output.write(str(s) + '\n')

with open("test.txt", "w") as output:
    for s in testlabel_names:
        output.write(str(s) + '\n')
```
